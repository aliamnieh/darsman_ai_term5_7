```python
# TODO: تبدیل به دیتاتایپ Dt_Customer و استخراج ویژگی

df['Dt_Customer']=pd.to_datetime(df['Dt_Customer'],format='%d-%m-%Y')
df['Dt_Customer']=df['Dt_Customer'].dt.date
print(df['Dt_Customer'].min())
print(df['Dt_Customer'].max())
```

<br>
<div dir="rtl">
    <font size="4">
        <li>
            یک ویژگی جدید به نام Customer_for  ایجاد کنید.
         ویژگی <strong>Customer_For</strong> نشان‌دهنده مدت‌زمانی است که هر مشتری از فروشگاه خرید می‌کند، بر اساس اختلاف تاریخ عضویت او با جدیدترین تاریخ ثبت‌شده در داده‌ها. </li> <li> برای این کار، ابتدا تاریخ عضویت مشتریان را به فرمت <strong>تاریخی</strong> تبدیل کنید. سپس جدیدترین تاریخ موجود در داده‌ها را مشخص کرده و اختلاف این دو تاریخ را برحسب روز محاسبه کنید.
        </li>
</font>
</div>

```python
# TODO: ایجاد ویژگی Customer_For

df['Customer_for']=df['Dt_Customer'].max()-df['Dt_Customer']
df['Customer_for']=df['Customer_for'].apply(lambda x:x.days)
```

```python
# TODO: حذف داده های پرت
def remove_outliers_iqr(df,columns):
    for column in columns:
        q1=np.percentile(df[column],20)
        q3=np.percentile(df[column],80)
        iqr=q3-q1
        lower_bound=q1-1.5*iqr
        upper_bound=q3+1.5*iqr
        df=df[(df[column]>=lower_bound) & (df[column]<=upper_bound)]
    return df
```

```python
# TODO: GMM پیدا کردن تعداد کلاستر های مناسب

from sklearn.mixture import GaussianMixture


n_range = range(2,11)

bic_score = []
aic_score = []

for n in n_range:
    gm = GaussianMixture(n_components=n, 
                         random_state=123, 
                         n_init=10)
    gm.fit(df)
    
    bic_score.append(gm.bic(df))
    aic_score.append(gm.aic(df))
    
fig, ax = plt.subplots(figsize=(12,8),nrows=1)
ax.plot(n_range, bic_score, '-o', color='orange')
ax.plot(n_range, aic_score, '-o', color='green')
ax.set(xlabel='Number of Clusters', ylabel='Score')
ax.set_xticks(n_range)
ax.set_title('BIC and AIC Scores Per Number Of Clusters')
```

```python
#TODO: GMM پیاده سازی

gmm = GaussianMixture(n_components=5, random_state=123, n_init=10)
preds = gmm.fit_predict(df)
sil_score=silhouette_score(df,preds)
print(f'silhouette score : {sil_score}')
print(f'davies bouldin score : {davies_bouldin_score(df,preds)}')
```
